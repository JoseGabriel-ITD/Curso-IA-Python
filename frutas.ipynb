```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Lista de colores para las clases
colors = ['Red', 'Orange', 'Blue', 'Green']

# Obtener las categorías únicas
fruit_names = df["fruit_name"].unique()

# Crear un mapeo entre clases y colores
color_dict = dict(zip(fruit_names, colors))

# Crear el gráfico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficar cada clase por separado
for fruit, color in color_dict.items():
    subset = df[df["fruit_name"] == fruit]
    ax.scatter(
        subset["height"],
        subset["width"],
        subset["color_score"],
        color=color,
        label=fruit,
        s=100  # Tamaño de los puntos
    )

# Etiquetas de los ejes
ax.set_xlabel("Height")
ax.set_ylabel("Width")
ax.set_zlabel("Color Score")

# Agregar la leyenda
ax.legend(title="Fruit Name")

# Mostrar el gráfico
plt.show()
```

## **Usando la librería seaborn**

**Explicación del Código**
Paleta de colores personalizada:

- Usamos sns.color_palette() para generar colores agradables asignados a cada clase (fruit_name).
hue_order asegura que el mapeo de colores sea consistente para todas las categorías.
Gráfico 3D con Matplotlib:

- Usamos *ax.scatter()* para graficar cada clase en el espacio 3D con las coordenadas height, width y color_score.
Leyenda:

- ax.legend() muestra la leyenda con los nombres de las frutas.
Estilo de Seaborn:

- Usamos sns.set(style="whitegrid") para aplicar el estilo de Seaborn al gráfico**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Aplicar un estilo de Seaborn
sns.set(style="whitegrid")

# Crear el gráfico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Asignar colores según las clases de frutas
hue_order = df["fruit_name"].unique()  # Obtener las categorías únicas
palette = sns.color_palette("husl", len(hue_order))  # Generar colores
color_dict = dict(zip(hue_order, palette))  # Mapear clases a colores

# Graficar cada clase por separado
for fruit, color in color_dict.items():
    subset = df[df["fruit_name"] == fruit]
    ax.scatter(
        subset["height"],
        subset["width"],
        subset["color_score"],
        color=color,
        label=fruit,
        s=100  # Tamaño de los puntos
    )

# Etiquetas de los ejes
ax.set_xlabel("Height")
ax.set_ylabel("Width")
ax.set_zlabel("Color Score")

# Agregar la leyenda
ax.set_title("Dataset Frutas")
ax.legend(title="Tipo de Fruta")

# Mostrar el gráfico
plt.show()
```
